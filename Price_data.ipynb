```python
#imports
import pandas as pd, shutil, os, time, glob
import numpy as np
from binance.client import Client
import json
import datetime
import requests
from statistics import mean
```

```python
#api connect
with open("credentials.json", "r") as f:
    creds = json.load(f)
client = Client(creds['API_KEY'], creds['SECRET_KEY'], tld='us')
```

```python
#functions
#convert to date_time
def convert_to_datetime(x):
    return datetime.datetime.fromtimestamp(x/1000)

#returns close price from klines 
def close_price(df_pair):
    close = client.get_historical_klines(df_pair, Client.KLINE_INTERVAL_1HOUR, "7 days ago UTC")
    c_p = pd.DataFrame(close, columns = ['ts_start', 'p_open','p_high', 'p_low', 'p_close','volume','ts_end','q_a_v','no_trades','volume_t','q_a_v_t','ignore'])
    c_p.ts_start = c_p.ts_start.apply(convert_to_datetime)
    c_p.ts_end = c_p.ts_end.apply(convert_to_datetime)
    c_p[["p_open", "p_high", "p_low","p_close", "volume"]] = c_p[["p_open", "p_high", "p_low","p_close", "volume"]].apply(pd.to_numeric)
    return c_p
```

```python
#pull pairs
df = pd.read_csv('Pairs/USD_pairs.csv')
coin_pairs = df.Pairs
print("The amount of coins chosen to observe: " + str(len(coin_pairs)))
```

```python
# Check that the amount of tickers isn't more than 2000
print("The amount of coins chosen to observe: " + str(len(coin_pairs)))
# These two lines remove the folder and then recreate it in order to remove old folder. Make sure you have created Folder the first time you run this.
shutil.rmtree("Price_data")
os.mkdir("Price_data")
#  These will do the same thing but for the folder jolding the RSI values. In progress
shutil.rmtree("RSI_data")
os.mkdir("RSI_data")
```

```python
#create blank df for closing prices

summary_closing_prices = pd.DataFrame()
```

```python
#add closing prices to df

for x in coin_pairs:
    temp_hist_data = close_price(str(x))
    temp_price = temp_hist_data['p_close']
    summary_closing_prices[x] = temp_price
```
